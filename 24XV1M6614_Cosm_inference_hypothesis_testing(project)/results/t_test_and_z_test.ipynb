```python
import pandas as pd
import numpy as np
from scipy import stats

# -------------------------------------
# Load your dataset
# -------------------------------------
df = pd.read_csv("CategoricalStudentData.csv")

# -------------------------------------
# Select two groups for comparison
# Example: Male vs Female Exam Scores
# -------------------------------------
group1 = df[df["Gender"] == "Male"]["Exam_Score"]
group2 = df[df["Gender"] == "Female"]["Exam_Score"]

# ============================================================
# 1Ô∏è‚É£ INDEPENDENT TWO-SAMPLE T-TEST
# ============================================================

t_stat, t_pvalue = stats.ttest_ind(group1, group2, equal_var=False)

print("üîπ T-Test Results")
print(f"T-statistic: {t_stat}")
print(f"P-value: {t_pvalue}\n")


# ============================================================
# 2Ô∏è‚É£ TWO-SAMPLE Z-TEST
# ============================================================
# Z-test = (mean1 ‚àí mean2) / SE
# SE = sqrt(œÉ1¬≤/n1 + œÉ2¬≤/n2)

mean_diff = group1.mean() - group2.mean()

se = np.sqrt(
    group1.var(ddof=1)/len(group1) +
    group2.var(ddof=1)/len(group2)
)

z_stat = mean_diff / se
z_pvalue = 2 * (1 - stats.norm.cdf(abs(z_stat)))  # two-tailed p-value

print("üîπ Z-Test Results")
print(f"Z-statistic: {z_stat}")
print(f"P-value: {z_pvalue}")
```
